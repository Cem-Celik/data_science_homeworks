There is no data in (1986-1991 and 2017,2019) period of years. So I will delete this particular data.

```python
for x in education.YEAR:
    if x < 1992 or x > 2016:
        education.drop(education.loc[education.YEAR == x].index,inplace = True)
```

```python
df = pd.DataFrame(education.groupby("YEAR").mean().reset_index())
df = df.set_index("YEAR")
```

There is no data in 1992 but we shouldn't lose other variables in 1992 so I change the variables in ENROLL same as 1993.

```python
education.ENROLL[education.loc[education.YEAR == 1992].index] = df.ENROLL.iloc[1]
df
```

GRADES_1_8_G and GRADES_9_12_G are missing between 1992-1996 :

```python
df_copy = education.copy()
a = education.GRADES_ALL_G[education.loc[education.YEAR == 1997].index].mean()
grades_12= education.GRADES_9_12_G[education.loc[education.YEAR == 1997].index].mean()
grades_8= education.GRADES_1_8_G[education.loc[education.YEAR == 1997].index].mean()
k_12 = grades_12/a
k_8 = grades_8/a
for x in df_copy.YEAR:
    if x < 1997:
        education.GRADES_1_8_G[education.loc[education.YEAR == x].index] = k_8 * education.GRADES_ALL_G[education.loc[education.YEAR == x].index].mean()
        education.GRADES_9_12_G[education.loc[education.YEAR == x].index] = k_12 * education.GRADES_ALL_G[education.loc[education.YEAR == x].index].mean()

```

```python
df = pd.DataFrame(education.groupby("YEAR").mean().reset_index())
df = df.set_index("YEAR")
df
```

**In general, I designed a algorithm in below, which is entering a mean value for every year for each variable in every column.**

```python
for x in education.select_dtypes(exclude='object')[2:]:
    if x != "YEAR":
        for i in range(0, len(education)):
            if pd.isnull(education[x].iloc[i]):
                education[x].iloc[i]=education.groupby('YEAR').mean()[x][education['YEAR'].iloc[i]]
```

```python
education.OTHER_EXPENDITURE[education.loc[education.YEAR == 1992].index] = df.OTHER_EXPENDITURE.iloc[1]
```
