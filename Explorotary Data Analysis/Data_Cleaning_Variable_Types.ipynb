```python
import pandas as pd
import numpy as np
movie = pd.read_csv("https://bootrain-lms-assets.s3.eu-central-1.amazonaws.com/bootrain-lms-static/datasets/movie.csv")
```

```python
display(movie)
```

```python
display(movie.budget.dropna())
```

"budget" is a numerical value, that can have infinite float values. So we can take this variable as a "Continous Variable" becuase it have infinite value and when we want to compare these values we can ratio these variables. So we can say this type of value are a "Ratio Variables".

```python
def make_smth(budget):
    if budget > 50000000:
        return "Huge budget"
    elif (budget <= 50000000) and (budget>15000000):
        return "Normal budget"
    elif (budget > 0 ) and (budget<=15000000): 
            return "Tiny budget"
    else:
        return "Unknown"
        
movie["budget_group"] = movie["budget"].apply(make_smth)

print(movie["budget_group"].value_counts())
```

**3- Is it possible to convert categorical variable into numerical one? If so, how can you make that?**

```python
def make_numeric(language):
    if language == "English":
        return 1
    else:
        return 0

movie["Is this movies English?"] = movie["language"].apply(make_numeric)

display(movie["Is this movies English?"])
```

I created a numeric value is change by categorical variable.
