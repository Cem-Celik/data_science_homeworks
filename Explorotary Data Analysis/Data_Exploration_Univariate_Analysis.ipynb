**5- You may have noticed that there are many missing values in the mathematics and reading notes. Complete the missing values with mean, median, and interpolation. Compare the impact of these techniques on the distribution of grades.**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
education = pd.read_csv("data/states_all.csv")

education
```

```python
df_mean = education.AVG_MATH_4_SCORE.fillna(education.AVG_MATH_4_SCORE.mean())
df_median = education.AVG_MATH_4_SCORE.fillna(education.AVG_MATH_4_SCORE.median())
df_interpolate = education.AVG_MATH_4_SCORE.interpolate(method="linear")

```

```python
plt.figure(figsize=(20,6))
plt.plot(df_mean)
plt.plot(df_median)
plt.plot(df_interpolate)
plt.legend(["df_mean","df_median","df_interpolate"])
```

```python
plt.figure(figsize=(20,6))
plt.hist(df_mean,alpha=.5)
plt.hist(df_median,alpha=.5)
plt.hist(df_interpolate,alpha=.5)
plt.legend(["df_mean","df_median","df_interpolate"])
```
