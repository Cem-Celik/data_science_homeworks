For Pandas assignments, you are going to use Titanic (train.csv) dataset. Download the dataset and load to a DataFrame.

```python
import pandas as pd
import numpy as np

train_df = pd.read_csv("data/train.csv", index_col = 0)
train_df
```

1- By using conditional selection methods, calculate the survival rate for all passengers under 30 years old.

```python
survived_under_30 = train_df[(train_df['Age']<30) & (train_df['Survived'] == 1)]

survival_rate = survived_under_30.shape[0]*100/train_df.shape[0]

print("Survival rate for all passengers under 30 years old is {:.2f}%".format(survival_rate))
```
