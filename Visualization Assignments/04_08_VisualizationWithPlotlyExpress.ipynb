In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

1. Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python


df_pl = df[(df["location"] == "Germany")|(df["location"] == "Spain")|
           (df["location"] == "Italy")|(df["location"] == "France")]
display(df_pl)
px.line(df_pl, x = df_pl.date, y = df_pl.total_deaths, title="Total Deaths for Four Countries", color="location")
```

2. Plot a bar plot with plotly express for average death number that compares continents.

```python
df_cont = df.groupby("continent").new_deaths.mean().reset_index()

fig = px.bar(data_frame = df_cont, x='continent', y='new_deaths',color='continent',
            labels={
                'continent':'Continents of the World',
                'new_deaths':'Average of deaths by day'
            }, title="Average of deaths by day of the World's Continents")
fig.show()
```

3. Plot a histogram for daily deaths for any country you choose. 

```python
df_his = df[df.loc[:,"location"]=="Turkey"]
display(df_his)
fig = px.histogram(df_his,x="new_deaths")
fig.show()
```
