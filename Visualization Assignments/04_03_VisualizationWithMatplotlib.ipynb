In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)

display(df)
```

General Data

```python
from datetime import datetime, timedelta

new_d = df["date"].sort_values()
last_day = new_d.iloc[-1]

d = last_day - timedelta(days=100) 

new_df = pd.DataFrame(df[df["location"].isin(["Spain","France","Germany","Italy"]) & (df['date']>d)])

display(new_df)
```

Spain-Italy-France-Germany last 100 days statics

```python
#title and axis' styles
title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 15 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }

plt.figure(figsize=(20,20))


plt.subplot(2,1,1)


plt.xticks(rotation=45,fontsize=10, c = "brown")
plt.xlabel('Time', fontdict = axis_style)
plt.ylabel("Number of Cases", fontdict = axis_style)
plt.title("Number of cases in all countries(Italy,France,Spain,Germany) per time", fontdict=title_style)


a = new_df[["location","total_cases","date"]]

germany = a[a["location"]=="Germany"]
plt.plot(germany["date"],germany["total_cases"],c="black",marker = "+")

france = a[a["location"]=="France"]
plt.plot(france["date"],france["total_cases"],c="blue",marker = "*", lw = 4)

spain = a[a["location"]=="Spain"]
plt.plot(spain["date"],spain["total_cases"],c="red" , marker = "o",lw = 2)

italy =  a[a["location"]=="Italy"]
plt.plot(italy["date"],italy["total_cases"],c="green",lw=3)
plt.legend(["GER","FRA","SPA","ITA"])
plt.subplot(2,1,2)
plt.title("Number of deaths for each countries(Italy,France,Spain,Germany) per time", fontdict=title_style)
plt.xticks(rotation=45,fontsize=10, c = "red")

plt.xlabel("Time",fontdict=axis_style)
plt.ylabel("Number of deaths in each countries", fontdict = axis_style)

a = new_df[["location","total_deaths","date"]]

germany = a[a["location"]=="Germany"]
plt.plot(germany["date"],germany["total_deaths"],c="black",marker = "+")

france = a[a["location"]=="France"]
plt.plot(france["date"],france["total_deaths"],c="blue",marker = "*", lw = 4)

spain = a[a["location"]=="Spain"]
plt.plot(spain["date"],spain["total_deaths"],c="red" , marker = "o",lw = 2)

italy =  a[a["location"]=="Italy"]
plt.plot(italy["date"],italy["total_deaths"],c="green",lw=3)

plt.legend(["GER","FRA","SPA","ITA"])

plt.show()
```
