1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python

df1 = df[(df.loc[:,"location"] == "Germany") | (df.loc[:,"location"]== "France") |
        (df.loc[:,"location"]== "Spain") | (df.loc[:,"location"]== "Italy") ]


plt.figure(figsize=(20,10))
plt.title("Total deaths in each countries",fontsize = 20)
plt.xlabel("Time",fontsize=20)
plt.ylabel("Total Deaths",fontsize=20)
sns.lineplot(x=df1.date,y=df1.total_deaths,hue=df1.location,data=df1)
plt.show()
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
df_new = df.groupby("continent").agg({"new_deaths":['mean']})

    
plt.figure(figsize=(20,10))
plt.title("Average Death Number of Continents",fontsize = 20)
plt.xlabel("Continents",fontsize=20)
plt.ylabel("Average Death Numbers",fontsize=20)
sns.barplot(x=df_new.index,y=df_new.iloc[:,0][:],data=df_new)
plt.show()
```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
df_hist = df[df["location"]=="Bangladesh"]

plt.figure(figsize=(30,20))

plt.subplot(2,2,1)
plt.title("Daily deaths in Bangladesh - Bin = 50 , KDE = True",fontsize = 20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Death Numbers",fontsize=20)
sns.distplot(df_hist.new_deaths,bins=50,kde=True)

plt.subplot(2,2,2)
plt.title("Daily deaths in Bangladesh - Bin = 50 , KDE = False",fontsize = 20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Death Numbers",fontsize=20)
sns.distplot(df_hist.new_deaths,bins=50,kde=False)

plt.subplot(2,2,3)
plt.title("Daily deaths in Bangladesh - Bin = 30 , KDE = True",fontsize = 20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Death Numbers",fontsize=20)
sns.distplot(df_hist.new_deaths,bins=30,kde=True)

plt.subplot(2,2,4)
plt.title("Daily deaths in Bangladesh - Bin = 30 , KDE = False",fontsize = 20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Death Numbers",fontsize=20)
sns.distplot(df_hist.new_deaths,bins=30,kde=False)

plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
df_last = df[(df["location"] == "Chile") | (df["location"] == "Peru")]


plt.figure(figsize=(15,6))

plt.subplot(1,3,1)
sns.boxplot(df_last.location,df_last.new_deaths,data=df_last)
plt.subplot(1,3,2)
sns.violinplot(x = "location", y = "new_deaths" , data=df_last)
plt.subplot(1,3,3)
sns.swarmplot(x = "location", y = "new_deaths" , data=df_last)
```
