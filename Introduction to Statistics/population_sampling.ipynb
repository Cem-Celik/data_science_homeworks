1- In this question, you test whether the central limit theorem works. You generate 1000 variables with two normal distributions. You can determine the mean and standard deviation of these variables yourself. All you have to do is generate the first variable 50 times and averaged it each time. Generate the second variable 1000 times and averages this variable each time. Then plot the histogram of the averages of the two variables. Which of the variables has a mean distribution closer to the normal distribution? Do you think the Central Limit Theorem seems to have worked?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
dist = np.append(np.random.normal(25,5,500),np.random.normal(75,5,500))

def take_sample(number,sub_plot):
    sample = []
    sample_means = []
    for x in range(number):
        sample.append(np.random.choice(dist,1)[0])
        sample_means.append(np.mean(sample))
    plt_sample(sample_means,number,sub_plot)
    print("Last sample error is =  {} ".format((np.mean(dist) - sample_means[-1])))
    
def plt_sample(sample_means,number, sub_plot):
    title_font = {'family' : 'arial', 'color' : 'darkred', 'weight': 'bold', 'size': 13}
    axis_font = {'family' : 'arial', 'color' : 'darkblue', 'weight' : 'bold', 'size' : 10}
    plt.subplot(sub_plot)
    plt.hlines(np.mean(dist),0,number,colors = 'r')
    plt.plot(sample_means)
    
    plt.title("Sample size = {}".format(number),fontdict = title_font)
    plt.xlabel("Sample size", fontdict = axis_font)
    plt.ylabel("Sample Mean", fontdict = axis_font)
    return plt

plt.figure(figsize=(20,8))
take_sample(50,121)
take_sample(1000,122)
```

Sometimes we may get better results from 50 samples but it is not correct when we looked from general. It is safe to use more samples size to get better and safer results.
